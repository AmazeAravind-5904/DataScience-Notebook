```python
def Talents(column):
    global df
    df[column].value_counts().sort_values(ascending=False)[:10].plot(kind="bar", figsize=(20,8))
    plt.xticks(rotation=0)
    plt.title("Top Ten {}".format(column))
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

```python
def Scatter(x, y):
    global df
    plt.figure(figsize=(20,8))
    plt.scatter(df[x], df[y], color="red")
    plt.xticks(rotation=0)
    plt.title("Scatter plot X:{} / Y:{}".format(x, y))
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

```python
Scatter("Rating","Votes")
```

```python
def Histo(column):
    global df
    plt.figure(figsize=(20,8))
    plt.hist(df[column],color="green")
    plt.xticks(rotation=0)
    plt.title("Histogram of {}".format(column))
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
